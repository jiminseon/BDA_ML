```python
import seaborn as sns
tt= sns.load_dataset('titanic')
```

```python
tt
```

```python
#남성의 평균 요금 
me_fa_mean=tt.query("""sex =='male'""")['fare'].mean()
```

```python
#여성의 평균 요금
fe_fa_mean=tt.query("""sex =='female'""")['fare'].mean()
```

```python
tt['female_fare_mean'] = 0
tt['male_fare_mean'] = 0

for i in range(len(tt)):
    # if문을 이용해서 남성과 여성의 값을 비교하고, 파생변수를 만들어서 그곳에 남성과 여성의 평균 fare값을 넣는 변수를 만든다.
    if tt.loc[i]['sex'] == 'female':
        tt['female_fare_mean'].loc[i] = fe_fa_mean
        
    elif tt.loc[i]['sex'] =='male':
        tt['male_fare_mean'].loc[i] = me_fa_mean
```

```python
tt[['sex','female_fare_mean','male_fare_mean']]
```

```python
#10대미만의 평균 요금 
ten_under_mean=tt.query(""" age < 10 """)['fare'].mean()
```

```python
#10대의 평균 요금 
ten_mean=tt.query(""" 10 <= age and age < 20 """)['fare'].mean()
```

```python
#20대의 평균 요금 
twenty_mean=tt.query(""" 20 <= age and age < 30 """)['fare'].mean()
```

```python
#30대의 평균 요금 
thirty_mean=tt.query(""" 30 <= age and age < 40 """)['fare'].mean()
```

```python
#그 외의 평균 요금 
others_mean=tt.query(""" 40 <= age """)['fare'].mean()
```

```python
tt['ten_under_fare_mean'] = 0
tt['ten_fare_mean'] = 0
tt['twenty_fare_mean'] = 0
tt['thirty_fare_mean'] = 0
tt['others_fare_mean'] = 0

for i in range(len(tt)):
    # if문을 이용해서 남성과 여성의 값을 비교하고, 파생변수를 만들어서 그곳에 남성과 여성의 평균 fare값을 넣는 변수를 만든다.
    if tt.loc[i]['age'] < 10:
        tt['ten_under_fare_mean'].loc[i] = ten_under_mean
        
    elif tt.loc[i]['age'] >= 10 and tt.loc[i]['age'] < 20:
        tt['ten_fare_mean'].loc[i] = ten_mean

    elif tt.loc[i]['age'] >= 20 and tt.loc[i]['age'] < 30:
        tt['twenty_fare_mean'].loc[i] = twenty_mean

    elif tt.loc[i]['age'] >= 30 and tt.loc[i]['age'] < 40:
        tt['thirty_fare_mean'].loc[i] = thirty_mean

    elif tt.loc[i]['age'] >= 40:
        tt['others_fare_mean'].loc[i] = others_mean
```

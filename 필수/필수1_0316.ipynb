```python
import seaborn as sns
tt = sns.load_dataset('titanic')
tt
```

```python
#1 여성이고 생존하고 지위가 제일 높은 사람
tt.query('''sex == "female" and survived == 1 and pclass == 1 ''')[['pclass','survived','sex']] 
```

```python
#2 여성이고 요금이 15보다 높고 지위가 제일 높은 사람
tt.query('''sex == "female" and fare > 15 and pclass == 1 ''')[['pclass','fare','sex']] 
```

```python
#3 남성이고 요금이 10보다 적고 지위가 제일 낮은 사람
tt.query('''sex == "male" and fare < 10 and pclass == 3 ''')[['pclass','fare','sex']] 
```

```python
#4 남성이고 요금이 10보다 적고 지위가 제일 높은 사람
tt.query('''sex == "male" and fare < 10 and pclass == 1 ''')[['pclass','fare','sex']]
```

```python
#5 혼자가 아니고 살아있으며 지위가 제일 낮은 사람
tt.query('''alone == False and alive == "yes" and pclass == 3 ''')[['alone','alive','pclass']] 
```

```python
#6 나이가 20살보다 적고 살아있지 않은 지위가 제일 높은 사람
tt.query('''age < 20 and alive == "no" and pclass == 1 ''')[['alone','alive','pclass']] 
```

```python
#7 여성이고 나이가 20살보다 적고 지위가 제일 낮은 사람
tt.query('''sex == "female" and age < 20 and pclass == 3 ''')[['alone','age','pclass']] 
```

```python
#7 남성이고 나이가 20살보다 적고 지위가 제일 낮은 사람
tt.query('''sex == "male" and age < 20 and pclass == 3 ''')[['sex','age','pclass']] 
```

```python
#7 여성이고 부모 자식은 없고 형제나 배우자가 명 이상인 사람
tt.query('''sex == "female" and parch == 0 and sibsp >= 2 ''')[['sex','parch','sibsp']] 
```

```python
#8 남성이고 부모 자식은 없고 형제나 배우자가 명 이상인 사람
tt.query('''sex == "male" and parch == 0 and sibsp >= 2 ''')[['sex','parch','sibsp']] 
```

```python
#9 남성이고 부모 자식은 없고 사망한 사람
tt.query('''sex == "male" and parch == 0 and survived == 0 ''')[['sex','parch','survived']] 
```

```python
#10 여성이고 부모 자식은 없고 사망한 사람
tt.query('''sex == "female" and parch == 0 and survived == 0 ''')[['sex','parch','survived']] 
```

```python
#11 여성이고 부모 자식이 1명이상이고 생존한 사람
tt.query('''sex == "female" and parch >= 1 and survived == 1 ''')[['sex','parch','survived']] 
```

```python
#12 최상위 지위고 부모 자식이 1명이상이고 생존한 사람
tt.query('''pclass == 1 and parch >= 1 and survived == 1 ''')[['pclass','parch','survived']] 
```

```python
#13 최상위 지위고 부모 자식이 1명이상이고 사망한 사람
tt.query('''pclass == 1 and parch >= 1 and survived == 0 ''')[['pclass','parch','survived']] 
```

```python
#14 여성이고 최상위 지위고 부모 자식이 1명이상이거나 사망한 사람
tt.query('''sex == "female" and pclass == 1 and parch >= 1 or survived == 0 ''')[['sex', 'parch','survived']] 
```

```python
#15 여성이고 최상위 지위고 부모 자식이 1명이상이고 사망한 사람
tt.query('''sex == "female" and pclass == 1 and parch >= 1 and survived == 0 ''')[['sex', 'parch','survived']] 
```
